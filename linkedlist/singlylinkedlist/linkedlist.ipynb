* WHAT IS LINKED LIST
# 
  A linked list is linear Data Structure consisting of a sequence of nodes.
 Where each node contains data and a reference (link) to the next node.
 the linked list are not stored in contiguous memory locations.
  They are accessed through a HEAD node and the final node pointing to NULL.
#  
*  KEY  CHARACTERISTICS.

A Node contains Data/Value --> which stores the value
  Next --> it stores the Reference of the next Node Address initially it is None.
  Head --> which is stores the first Node in it.
  Tail --> which stores the Last Node.
 
*  IN PYTHON THE CREATION OF LINKED LIST IS POSSIBLE WITH THE HELP OF THE CLASS
 The Node data is defined inside the constructor.
  
* TYPES OF LINKED LIST
  1. Singly Linked List
  2. Doubly Linked List
  3. Circular Linked List
  
*  COMMON APPLICATIONS
  
 Linked lists are used for implementing stacks, queues, and graphs.
