# ARRAYS

**Arrays** are the major basic data structure in the python

These arrays are used to store the **same data type** under it.

The length of the array is **variable length** because it can change its length

In python we can create array with the help of using the **list**.

We consider **list as an array**

The list contains many **built-in methods**

---

## **Insertion Methods**

### **Append** 
- It will add the single value or collection data type as a single value and it will add at the **last**.

### **Extend**
- It will add the **only collection data type** and extract each and every elements and append them into the **last**

### **Insert**
- It will insert the value based on the **index position**.
- If the index position is large then it will add at the **last**
- If the index position is small it will add at the **first**

---

## **Deletion Methods**

### **Pop**
- It is used to delete the **last element** one at a time.
- It will take the **one non-mandatory argument** which is index position.

### **Remove**
- It is used to delete the element based on the **value**.
- If the value is not present it will throw an **error**.

---

## **TYPES OF ARRAYS**

The arrays are classified into **two types**:

1. **ONE DIMENSIONAL ARRAY**
2. **TWO DIMENSIONAL ARRAY**

---

### **1. One Dimensional Array**

These are the arrays which stores the elements in the **single sequence**. Each elements are accessed by using the **single index** which is starting from **0**.

**Example:**
```python
arr = [1, 2, 3, 4, 5, 6, 8]
```

### **2. Two Dimensional Array**

These are the arrays which stores the elements in the form of **row and column** similar to grid or matrix. To access the particular elements it always requires the **two index values** - one is for **row** and another is for **column**.

**Example:**
```python
arr = [[1, 2, 3],
       [5, 9, 7],
       [44, 53, 23]]
```
