# STRINGS

**Strings** are used to store the **individual elements** in the **sequence order**

The strings are **immutable** data types

Strings are the **fixed length** collection data type

To perform any **manipulation operations** on the strings we use the **indexing** and **slicing**

Strings are stored in the form of **single quote mark** (' ') or **double quote marks** (" ")

After creating the string, it is **not possible to change** the values

String is an **ordered collection** data type

---

## **String Built-in Methods**

String contains many **built-in methods**. Majorly used are:

- **lower()** - Converts string to lowercase
- **upper()** - Converts string to uppercase  
- **split()** - Splits string into list
- **replace()** - Replaces part of string
- **join()** - Joins list elements into string
- **format()** - Formats the string
- **len()** - Returns length of string
- **count()** - Counts occurrences of substring
- **isalnum()** - Checks if string is alphanumeric

---

## **Examples**

```python
# String examples
text = "Hello World"

print(text.lower())     # hello world
print(text.upper())     # HELLO WORLD
print(text.split())     # ['Hello', 'World']
print(len(text))        # 11
print(text.count('l'))  # 3
```
